```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 5000)
```

```python
df=pd.read_csv('./inputs/creditcarddata/creditcard_2023.csv', index_col='id')
```

```python
display(df.head())
```

```python
display(df.info())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns = ['Class']),
    df['Class'],
    stratify = df['Class'],
    shuffle = True,
    random_state = 0,
)
```

```python
print(X_train.shape, X_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
display(X_train.info())
```

```python
display(X_train.describe().T)
```

```python
display(y_train.value_counts())
```

```python
fig, axes = plt.subplots(X_train.shape[1], 2, figsize = (16, 4 * X_train.shape[1]))
for i in range(X_train.shape[1]):
    x = X_train.iloc[:, i]
    sns.histplot(x = x, bins = 30, kde = True, hue = y_train, ax = axes[i, 0])
    sns.boxplot(x = x, y = y_train, orient = 'h', ax = axes[i, 1])
fig.tight_layout();
```

```python
corr_train = pd.concat([X_train, y_train], axis = 1).corr()
```

```python
plt.figure(figsize = (16, 12))
sns.heatmap(corr_train, vmin = -1, vmax = 1, cmap = 'coolwarm', annot = True);
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
from sklearn.linear_model import LogisticRegression

logistic_pred = LogisticRegression(
    random_state = 0,
    max_iter = 1000,
).fit(X_train, y_train).predict(X_test)
```

```python
print(classification_report(y_test, logistic_pred, digits = 4))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, logistic_pred, values_format = 'd');
```

## Power Transform

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PowerTransformer
```

```python
power_lin_pred = make_pipeline(
    PowerTransformer(),
    LogisticRegression(
        random_state = 0,
        max_iter = 1000,
    )
).fit(X_train, y_train).predict(X_test)
```

```python
print(classification_report(y_test, power_lin_pred, digits = 4))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, power_lin_pred, values_format = 'd');
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb_pred = HistGradientBoostingClassifier(
    random_state = 0,
).fit(X_train, y_train).predict(X_test)
```

```python
print(classification_report(y_test, hgb_pred, digits = 4))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, hgb_pred, values_format = 'd');
```
